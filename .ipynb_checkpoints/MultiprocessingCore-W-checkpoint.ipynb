## Pool Process Function
This function will be used to map a function on data to the specified number of processes.

Note: I have commented out the print statements and the time measurement to increase the accuracy of my timing function.

```python
# A function for timing a job that uses a pool of processes.
#  f is a function that takes a single argument
#  data is an array of arguments on which f will be mapped
#  pool_size is the number of processes in the pool. 
def pool_process(f, data, pool_size = multiprocess.cpu_count()):
    if __name__ == '__main__':
        with multiprocess.get_context('spawn').Pool(processes=pool_size) as pool:
#             tp1 = time.time()
#             print(data, pool_size)
            result = pool.map(f, data)       # map f to the data using the Pool of processes to do the work 
            pool.close() # No more processes
            pool.join()  # Wait for the pool processing to complete. 
#             print("Results", result)
#             print("Overall Time:", int(time.time()-tp1))
        
```

## Obtaining a suitable Prime Number
The below function returns a list of prime numbers. The list will return the same prime number as each element in the list. This will be useful to ensure that the analysis over a number of cores in decently comparable. 

```python
def get_primes_list(array_lengths, prime_range):
    """
    Get list of primes based on array lengths
    """
    primes_list = { i : [] for i in array_lengths }
    primeNumber = randprime(prime_range[0], prime_range[1])
    primeList = []
    for i in primes_list:
        for j in range(i):
            primes_list[i].append(primeNumber)
    return primes_list
```

## Timing the function
The below function is used to time how long processing takes for each number of cores. It takes a number of parameters including the list of prime numbers which will be tested on along with the number of iterations to do and the number of cores to analyse.

This function could be run on a PC with more/less cores than mine by simply changing some of the parameters.

```python
# average case timer function
def timer_func(primes, number_of_iterations=10, number_of_cores=4):
    """
    Function to time 3 search algorithms: Linear, Binary Iterative & Recursive
    Saves data to dataframes which are then returned
    """
    df_list = []
    for i in range(0,number_of_iterations):
        print(("*"*20)+f"Iteration Number {i}"+("*"*20))
        for core in range(1, number_of_cores+1):
            print(("-"*10)+f"Core Number {core}, Iteration Number {i}"+("-"*10))
            df_list.append(pd.DataFrame(index=range(0,number_of_iterations), columns=primes.keys()))
            times = []
            for j in primes:
                start_time = timer()
                pool_process(check_prime, primes[j], core)
                elapsed_time = timer() - start_time
                times.append(elapsed_time)
            df_list[core-1].loc[i] = times
    return df_list
```

## Testing the function
This functions simply groups together getting the list of prime numbers and timing it.

```python
def testing_func(array_lengths = [1, 5, 10, 25, 50, 100, 150, 200, 250, 300], prime_range = [100000, 102000], iterations=20, cores=4):
    """
    Initiate the testing with a number of inputs
    """
    primes_list = get_primes_list(array_lengths, prime_range)
    df_list = timer_func(primes_list, iterations, cores)
    return df_list
```

## Plotting the Results
The below function is used to render the plots after the times have been recorded. It is calibrated to work with the functions define above.

```python
def get_plot(df, title_add,os="Windows OS", cores=4, array_lengths=None, figname="base"):
    labels = list(map(str, array_lengths))

    x = np.arange(len(labels))  # the label locations
    width = 0.10 # the width of the bars
    plt.rc('font', size=20)
    fig, ax = plt.subplots(figsize=(15,15))
    ax.set_axisbelow(True)
    ax.yaxis.grid(color='gray', linestyle='dashed')
    if cores >= 1:
        rects1 = ax.bar(x-.15, df[0].mean().T, width, label='1 Core', color=['#a9a9a9'])
    if cores >= 2:
        rects2 = ax.bar(x-.05, df[1].mean().T, width, label='2 Core', color=['#4363d8'][::-1])
    if cores >= 3:
        rects3 = ax.bar(x+.05, df[2].mean().T, width, label='3 Core', color=['#ffe119'])
    if cores >= 4:
        rects4 = ax.bar(x+.15, df[3].mean().T, width, label='4 Core', color=['#000000'][::-1])

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Time in Seconds')
    plt.xlabel("Prime Array Length")
    ax.set_title(os + ' - MP Avg. Times ' + title_add)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, .99),
          ncol=2, fancybox=True, shadow=True)

    high_point_x = []
    high_point_y = []    
    
    for i in range(len(labels)):
        high_point_x.append(labels[i])
        for core in range(cores-1):
            if core-1 < len(high_point_y):
                high_point_y.append([])
#                 print(high_point_y)
            diff = (df[core+1].mean().T[array_lengths[i]] - df[0].mean().T[array_lengths[i]])/ df[0].mean().T[array_lengths[i]]
            high_point_y[core].append(diff)
    
    
    color_list = ["#a9a9a9", "#4363d8", "#ffe119", "#000000"]
#     print(high_point_x, high_point_y)
    ax2 = ax.twinx()
    for core in range(cores-1):
        if core == 1:
            continue
        rects3 = ax2.plot(high_point_x,high_point_y[core],marker='o', color=color_list[core+1], label='Trend Line')
        for i,j in zip(high_point_x,high_point_y[core]):
            value = np.round(j,2)
            if value > 0:
                value= "+" + str("{:.0%}".format(np.round(j,2)))
            else:
                value= str("{:.0%}".format(np.round(j,2)))
            ax2.annotate(value,xy=(i,j+0.01), path_effects=[pe.withStroke(linewidth=4, foreground="white")])


    fig.tight_layout()
    
    plt.savefig("images/" + figname + ".png")
    plt.show()
```

## Saving the results to Pickle files

```python
def save_pickle_list(df_list, name):
    """
    Save list of dataframes to pickle file
    """
    file = open(f'pickles/{name}.pickle', 'ab') 
      
    # source, destination 
    pickle.dump(df_list, file)                      
    file.close() 
    
def load_pickle_list(name):
    """
    Load pickle file to variable
    """
    with open(f'pickles/{name}.pickle', 'rb') as f:
        x = pickle.load(f)
    return x
```

## Comparison with Amdahl's Law
Amdahl's law is a formula which gives the theoretical speedup in latency of the execution of a task at fixed workload that can be expected of a system whose resources are improved.

By defining a function which gives us the returned value from the Amdahl's law function we can use this to compare this against our actual data.

Unfortunately, this was initially made under the pretense that in the case of, for example, a prime number list of length 4 this would be 75% parallelisable as we can parallelise it down to 1/4 of the original running time. This is in fact not exactly how Amdahl's law works. Amdahl's law works much better in cases where there is a portion of the program which is Serial and a portion which is parallelisable. In our case, the prime number function as it is defined is 100% parallelisable as there is no aspect which we are not parellelising. Nonetheless, the function below could be used in the future, for analysis of a program which encorporates both Serial and Parallel portions.

```python
def amdahls_law(n, p):
    return 1/((1-p)+(p/n))
```

```python
def amdahls_law_plot(df, cores=4, title="" ):
    colors = list(mcolors.TABLEAU_COLORS)
    df_concat = pd.DataFrame(columns=[1,2,3,4])
    for i in range(4):
        df_concat[i+1] = df[i].mean()

    linestyle_str = [
     ('solid', 'solid'),      # Same as (0, ()) or '-'
     ('dotted', 'dotted'),    # Same as (0, (1, 1)) or '.'
     ('dashed', 'dashed'),    # Same as '--'
     ('dashdot', 'dashdot')] 
    speedups = [[1] for i in range(5)]
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    x = np.linspace(1,4)
    y = amdahls_law(x, 0.75)
    for i in range(1,5):
        y = amdahls_law(x, 1-(1/i))
#         print((1-1/i))
        for j in range(1,4):
            speedups[i].append(df_concat.loc[i][j]/df_concat.loc[i][j+1])
        plt.plot(x,y, colors[i-1], linestyle=linestyle_str[0][0])
        plt.plot([1,2,3,4],speedups[i], colors[i-1], linestyle=linestyle_str[2][0])
    plt.title("Amdahls Law")
    plt.legend(['p=0', '1 prime','p=0.5', '2 prime','p=2/3', '3 primes','p=0.75', '4 primes']);
    plt.xlabel("Number of Cores")
    plt.ylabel("Speedup")
    plt.xticks([1,2,3,4])
    plt.savefig(f"images/Amdahls_Law_{title}.png")
    plt.show()
```

```python
amdahls_law_plot(df_large_primes_short_array, cores=4, title="Windows")
```

Please note that the plot above does not accurately reflect amdahl's law in this case.
